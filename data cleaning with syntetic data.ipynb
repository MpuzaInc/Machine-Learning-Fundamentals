## What is Imputation?
### Imputation means filling in missing values in your dataset instead of removing them. This helps preserve data and avoid losing important patterns during analysis or model training.

## Use imputation when:
### You have missing values that are not too many (e.g., < 30% of a column).
### Dropping rows would result in losing too much useful data.
### The missing data seems to follow a pattern or is missing at random.

## When NOT to Impute:
### If the missing data has a meaning (e.g., test not taken, product never sold).
### If the column is mostly empty (e.g., 80%+ missing), consider dropping it.

## Imputation Type	Use This When...	Best For	Example
### Mean (.mean())	: Data is numerical and normally distributed (no big outliers)
### Median (.median())	: Data is numerical but has outliers or is skewed
### Mode (.mode())	: Data is categorical (not numbers)

## The KDE (Kernel Density Estimate) curve overlaid on the histogram gives a good visual cue. A bell-shaped curve is a good sign of normality. here this curve showing that it is normaly distributed cz is bell shapped curve

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(dataset['Salary'], kde=True)
plt.title("Histogram with KDE")
plt.show()
```

## Since p = 0.999 > 0.05, we fail to reject the null hypothesis that the data comes from a normal distribution.

## Interpretation: The data in Salary is probably normally distributed.

```python
from scipy.stats import kstest, norm

isNormalDistribution = dataset['Salary'].dropna()
stat, p = kstest((isNormalDistribution - isNormalDistribution.mean()) / isNormalDistribution.std(), 'norm')
print('K-S Test: stat=%.3f, p=%.3f' % (stat, p))

if p > 0.05:
    print("Probably normal")
else:
    print("Probably not normal")
```

## we plot this column with boxplot to check if it contain outliers, we luckly see that it has no outliers

```python
plt.boxplot(x=dataset['Salary'])
plt.title('Boxplot of Salary')
plt.show()
```

## from the above all reason, it is approved that we may impute this salary missing value using "Mean", because it is normally distributed and have no outliers

```python
dataset.loc[:,"Salary"] = dataset["Salary"].fillna(dataset["Salary"].mean())
```

## HANDLING OUTLIERS ON Age Column Using Quantile where we set upper bound and lower bound using statistical approach

```python
Q1 = dataset["Age"].quantile(0.25)
Q3 = dataset["Age"].quantile(0.75)
IQR = Q3 - Q1
lowerb = Q1 - 1.5*IQR
upperb = Q3 + 1.5*IQR
```

```python
outliers = dataset[(dataset["Age"] < lowerb) | (dataset["Age"] > upperb )]
outliers
```

## Why Not Just Keep the Month as 1â€“12?
#### If you keep the DoB_month as a plain number (1 for January, 12 for December), a machine learning model will think:

#### December (12) is far away from January (1),

#### And March (3) is closer to December than January is.

## But that's wrong!
#### December and January are right next to each other on the calendar. So a simple number does not reflect the true relationship.

#### This is where cyclical encoding comes in.

## Why Use Sine and Cosine?
#### Using sin() and cos() helps encode the cyclic pattern on a circle instead of a line.

#### Think of the months placed around a clock (a circle):

#### January (1) = 30Â°

#### April (4) = 120Â°

#### July (7) = 210Â°

#### December (12) = 360Â° (or back to 0Â°)

#### This way, December and January are close again, and the model can learn cyclical proximity.

```python
dobMonthIndex = dataset.columns.get_loc('DoB Months')
```

```python
dobMonthIndex
```

```python
dataset.insert(dobMonthIndex + 1, 'DoB Month Sin', np.round(np.sin(2 * np.pi * dataset['DoB Months'] / 12), 3))
```

```python
dataset.insert(dobMonthIndex + 2, 'DoB Month Cos', np.round(np.cos(2 * np.pi * dataset['DoB Months'] / 12), 3))
```
